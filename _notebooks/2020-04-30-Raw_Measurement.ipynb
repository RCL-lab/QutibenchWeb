```python
#hide
#imagenet
for df in dataframes:
    df = df.drop(columns = ['NN_Topology', 'HWType', 'Precision','level','GOPS',
                        'hw_precision_net_prun','norm-lat-comp','quant_model', 'PruningFactor'])
   
    df = df[['hw_quant_prun_net', 'Op mode', 'batch/thread/stream', 'lat-comp', 'fps-system', 'fps-comp',
       'tp-system', 'tp-comp', 'top1', 'top5 [%]', 'Base_Pwr_W', 'Idle_Pwr_W',
       'Full_Pwr_W']]
    
    df.columns = ['hw_quant_prun_net', 'Op mode', 'batch', 'lat-comp', 'fps-system', 'fps-comp',
       'tp-system', 'tp-comp', 'top1', 'top5 [%]', 'Base_Pwr_W', 'Idle_Pwr_W',
       'Full_Pwr_W']
    path = 'data/latex/latex_' + df['hw_quant_prun_net'].iloc[0] + '.txt'
    df.to_latex(buf=path,
                header = ['hw_quant_prun_net ', 'Op mode', 'batch', 'lat-comp',
       'fps-system', 'fps-comp', 'tp-system', 'tp-comp', 'top1', 'top5 [%]',
       'Base_Pwr_W', 'Idle_Pwr_W', 'Full_Pwr_W'],
                column_format= '|r|c|l|r|c|l|r|c|l|r|c|l|r|c|l|r|c|l|r|c|l|',
                #column_format='lp{1.8cm}p{1.8cm}p{1.8cm}p{1.8cm}p{1.8cm}p{1.8cm}',
                na_rep='na',
                bold_rows='True',
                index=False)
```
